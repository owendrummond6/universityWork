The restaurant menu consists of a selection of pizzas, but this prototype application focuses on letting a customer order their own pizza (£12.99), selecting 2 out of the following toppings: 

-	Mozzarella
-	Pepperoni
-	Ham
-	Pineapple
-	Olives

Besides pizzas, the customer can also order a portion of garlic bread (£4.99) and also a drink (£2.99) - various drinks are available but the customer will help themselves to these via a self-service machine (so the app should just display ‘Drink’ or ‘Beverage’ rather than specific drinks).

```python
## using Object Oriented Programming to create our Menu Items

class Pizza:
    """
    Represents a customer's pizza order,
    including the selected toppings and price,
    as well as the confirm_order() method for pizza.
    """
    def __init__(self, toppings,): ## initialise with a toppings variable
        self.toppings = toppings
        self.price = 12.99 ## set a pizza.price variable to 12.99 to be used in the running_total logic

    def confirm_order(self):
        toppings_text = " and ".join(self.toppings) ## concatenate the toppings dividied by 'and'; double toppings logic embedded below within select_toppings() function
        print(f"Thank you, a {toppings_text} pizza has been added and is on its way!\n") ## print confirmation text
        

class GarlicBread:
    """
    Represents the order of a Garlic Bread,
    including the price as well as the
    confirm_order method for garlic bread.
    """
    def __init__(self):
        self.price = 4.99 ## set the GarlicBread.price variable for used in running_total logic

    def confirm_order(self):
        print("\nThank you, a garlic bread has been added and is on its way!\n") ## print confirmation text


class Drink:
    """
    Represents the order of a Drink,
    including the price and the
    confirm_order method for a drink.
    """
    def __init__(self):
        self.price = 2.99 ## set the Drink.price variable

    def confirm_order(self):
        print("\nThank you, a drink has been added.") ## printing the confirmation text
        print("You will be brought a drinking glass for the self-service machine!\n") ## printing the confirmation text
```

```python
def select_toppings():
    """
    Create a select_toppings() function for use in the follow-through menu of selecting a pizza;
    a dictionary is created for the toppings, assigning a Key-Value pair.
    The Keys represent the user input, and the values represent the full topping names.
    """
    topping_mapping = {
        "A": "Pepperoni", 
        "B": "Mozzarella",
        "C": "Ham",
        "D": "Pineapple",
        "E": "Olives"
    }

    ## Allow for User to select their desired toppings

    while True:
        """
        Create an infinite loop to ask the user for input until valid toppings are selected;
        contains the error-handling of cases such as inputs: A,A,A , Z,Z , or C only.
        """
        topping_choices = input(
            "\nPlease select any 2 toppings:\n"
            "A - Pepperoni\n"
            "B - Mozzarella\n"
            "C - Ham\n"
            "D - Pineapple\n"
            "E - Olives\n"
            "Enter 2 choices separated by commas, e.g. A,B: "
        ).strip().upper().split(",") ## format management of user input - remove whitespace, uppercase letters, separates the two selections into a list

        ## Further formatting of spaces
        topping_choices = list(map(lambda x: x.strip(), topping_choices)) ## removes whitespace

        # Check exactly 2 toppings
        if len(topping_choices) != 2: ## checks if the length of the user's input is 2
            print("Please select exactly 2 toppings.") ## if is isn't length 2, ask for users to re-input
            continue ## else continue with select_toppings()

        # Check both are valid and found in topping_mapping dictionary
        if topping_choices[0] not in topping_mapping or topping_choices[1] not in topping_mapping: ## checks both characters 1 and 2 are keys found within topping_mapping
            print("Please only choose valid topping options: A, B, C, D or E.") ## displays error message and user is asked to re-enter their choices
            continue ## else continue with select_toppings()

        # Convert letters into topping names
        selected_toppings = list(
            map(lambda x: topping_mapping[x], topping_choices) ## lambda function is used to convert user input (key) to topping name (value), i.e ["A", "B"] becomes ["Pepperoni", "Mozzarella"]
        )

        # Handling double toppings
        if selected_toppings[0] == selected_toppings[1]: ## checks to see if both toppings are the same, e.g ["Pepperoni", "Pepperoni"]
            selected_toppings[0] = f"Double {selected_toppings[0]}" ## update the first entry to become ["Double Pepperoni"], so now the list is ["Double Pepperoni", "Pepperoni"]
            selected_toppings.remove(selected_toppings[1]) ## remove the duplicate entry (the second item in the list), now selected_toppings becomes ["Double Pepperoni"]

        return selected_toppings ## return the list of toppings
```

```python
# Display the price
running_total = 0 ## create a running_total variable for displaying the end price - starts at 0

# Create the Main Menu
def query_user_input(running_total): ## creates the main menu function with running_total to show available products and manage the user's bill
    while True: ## create infinite loop to repeatedly prompt the user for input until they are finished (checkout or exit)

        chosenItem = input( ## displays the main menu option
            "Welcome to Uncle Enzo's Pizza!\n"
            "Please select whether you would like to order, or checkout:\n"
            "A - Pizza - £12.99\n"
            "B - Garlic Bread - £4.99\n"
            "C - Drink - £2.99\n"
            "D - Checkout\n"
            "X - Exit\n"
            "Please select an option using A, B, C, D or X: "
        ).strip().upper() ## input standardisation

        # Pizza Ordering Process
        if chosenItem == "A": ## check if their input is "A" for pizza
            toppings = select_toppings() ## display the pizza toppings menu
            pizza = Pizza(toppings) ## a Pizza object is created with the user's toppings
            pizza.confirm_order() ## the order confirmation message is displayed
            running_total += pizza.price ## the pizza's price is added to the running_total variable

        # Garlic Bread Ordering Process
        elif chosenItem == "B": ## check if the input is "B" for pizza
            garlic_bread = GarlicBread() ## create the Garlic Bread object  
            garlic_bread.confirm_order() ## the order confirmation message is displayed
            running_total += garlic_bread.price ## the garlic bread's price is added on to the running_total variable
 
        # Drink Ordering Process
        elif chosenItem == "C": ## check if the input is "C" for drink
            drink = Drink() ## create Drink object
            drink.confirm_order() ## the order confirmation message is displayed
            running_total += drink.price ## the drink's price is displayed

        # Checkout Process
        elif chosenItem == "D": ## check if the input is "D" for checkout 
            if running_total == 0: ## check if the user has ordered nothing (their bill is £0.00)
                print("Thank you for visiting Uncle Enzo's Pizza! Sorry you couldn't find anything you were looking for!") ## display checkout message
                break ## stops code
            else:
                print(f"\nThank you for visiting Uncle Enzo's Pizza. Your bill is £{running_total:.2f}, and your order will be on its way shortly!") ## products have been ordered, display the confirmation message and final bill
                break ## stops code
        elif chosenItem == "X": ## check if the input is "X" for exit
            print("\nThank you for visiting Uncle Enzo's Pizza!") ## display the exit message
            break ## stops code

        else:
            print("\nPlease select a valid option: A, B, C, D or X.\n") ## prompt the user to input a valid input character
```

```python
query_user_input(running_total) ## run code
```
