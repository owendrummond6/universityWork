The restaurant menu consists of a selection of pizzas, but this prototype application focuses on letting a customer order their own pizza (£12.99), selecting 2 out of the following toppings: 

-	Mozzarella
-	Pepperoni
-	Ham
-	Pineapple
-	Olives

Besides pizzas, the customer can also order a portion of garlic bread (£4.99) and also a drink (£2.99) - various drinks are available but the customer will help themselves to these via a self-service machine (so the app should just display ‘Drink’ or ‘Beverage’ rather than specific drinks).

```python
class Pizza:
    def __init__(self, toppings,):
        self.toppings = toppings

    def confirm_order(self):
        self.price = 12.99
        print("\nThank you. Your pizza has been ordered with:")
        for topping in self.toppings:
            print(f"- {topping}")
        print("Your pizza is on its way!\n")


class GarlicBread:
    def confirm_order(self):
        self.price = 4.99
        print("\nThank you. You ordered Garlic Bread. It is on its way!\n")


class Drink:
    def confirm_order(self):
        self.price = 2.99
        print("\nThank you. You ordered a drink.")
        print("You will be brought a drinking glass for the self-service machine!\n")#
```

```python
def select_toppings():
    topping_mapping = {
        "A": "Pepperoni",
        "B": "Mozzarella",
        "C": "Ham",
        "D": "Pineapple",
        "E": "Olives"
    }

    while True:
        topping_choices = input(
            "\nPlease select any 2 toppings:\n"
            "A - Pepperoni\n"
            "B - Mozzarella\n"
            "C - Ham\n"
            "D - Pineapple\n"
            "E - Olives\n"
            "Enter 2 choices separated by commas, e.g. A,B: "
        ).strip().upper().split(",")

        # Clean spaces
        topping_choices = list(map(lambda x: x.strip(), topping_choices))

        # Check exactly 2 toppings
        if len(topping_choices) != 2:
            print("Please select exactly 2 toppings.")
            continue

        # Check both are valid
        if topping_choices[0] not in topping_mapping or topping_choices[1] not in topping_mapping:
            print("Please only choose valid topping options: A, B, C, D or E.")
            continue

        # Convert letters into topping names
        selected_toppings = list(
            map(lambda x: topping_mapping[x], topping_choices)
        )

        print(selected_toppings)

        if selected_toppings[0] == selected_toppings[1]:
            selected_toppings[0] = f"Double {selected_toppings[0]}"
            selected_toppings.remove(selected_toppings[1])

        return selected_toppings
```

```python
running_total = 0
print(running_total)

def queryUserinput(running_total):
    while True:
        chosenItem = input(
            "Welcome to Uncle Enzo's Pizza!\n"
            "Please select whether you would like to order, or checkout:\n"
            "A - Pizza\n"
            "B - Garlic Bread\n"
            "C - Drink\n"
            "D - Checkout\n"
            "Please select an option using A, B, C or D: "
        ).strip().upper()

        if chosenItem == "A":
            toppings = select_toppings()
            pizza = Pizza(toppings)
            pizza.confirm_order()
            running_total += pizza.price

        elif chosenItem == "B":
            garlic_bread = GarlicBread()
            garlic_bread.confirm_order()
            running_total += garlic_bread.price 

        elif chosenItem == "C":
            drink = Drink()
            drink.confirm_order()
            running_total += drink.price

        elif chosenItem == "D":
            print(f"\nThank you for visiting Uncle Enzo's Pizza. Your bill is {running_total}!")
            break

        else:
            print("\nPlease select a valid option: A, B, C or D.\n")
```

```python
queryUserinput(running_total)
```
